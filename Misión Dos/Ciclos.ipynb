```python
def comparar_numeros(inicio,fin,divisor):
  for i in range(inicio,fin):
    if i % divisor == 0:
      continue
  print(f"El exponente de {i} es: {i**2}")
# Posicion en una lista[0,1,2]
inicio = [1,2,3]
fin = [10,100,200]
divisor = [2,3,4]

comparar_numeros(inicio[1],fin[0],divisor[2])

```

```python
from timeit import default_timer as timer
from datetime import timedelta
cadena = " Me encanta programar"

def contar_letras(cadena,letra):
    contador = 0
    for i in cadena.upper():
        if i == letra.upper():
            contador += 1
    print(f"Veces en las que esta '{letra}' en la cadena = {contador}")

start = timer()
contar_letras(cadena,"m")
end = timer()

print(timedelta(seconds = end - start))
```

```python
from timeit import default_timer as timer
from datetime import timedelta
cadena = " Me encanta programar"

def contar_letras(cadena,letra):
    for j in letra:
        contador = 0
        for i in cadena.upper():
            if i == j.upper():
                contador += 1
        print(f"Veces en las que esta '{j}' en la cadena = {contador}")

start = timer()
contar_letras(cadena,"map")
end = timer()

print(timedelta(seconds = end - start))
```

```python
from timeit import default_timer as timer
from datetime import timedelta
cadena = " Me encanta programar"

def contar_letras(cadena,letra):
    contador = 0
    for i in cadena.upper():
        if i == letra.upper():
            contador += 1
    print(f"Veces en las que esta '{letra}' en la cadena = {contador}")

start = timer()
contar_letras(cadena,"m")
contar_letras(cadena,"a")
contar_letras(cadena,"p")
end = timer()

print(timedelta(seconds = end - start))
```
